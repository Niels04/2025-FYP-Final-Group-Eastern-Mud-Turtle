```python
from sklearn.model_selection import GroupShuffleSplit, train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
import pandas as pd
import numpy as np
```

```python
#split data into training and test set
md = pd.read_csv("../data/metadata.csv")
md.head()
```

```python
#classifiers(x) = itch, bleed, elevation
#diagnostic(y) 
md = md[["diagnostic","bleed","itch","elevation","grew","hurt","changed"]]
md["is_melanoma"] = (md["diagnostic"] == "MEL").astype(int)
md["bleed"] = md["bleed"].astype(bool).astype(int)
md["itch"] = md["itch"].astype(bool).astype(int)
md['elevation'] = md['elevation'].astype(bool).astype(int)

x = md[["bleed", "itch", "elevation"]]
y = md["is_melanoma"]

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=42)
```

```python
#the model
model = LogisticRegression(class_weight="balanced")
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
```

```python
#test the model;
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
def logistic_model(metadata_dir): #give it classified.csv as input
    cd = pd.read_csv(metadata_dir)

    #classifiers A,B,C
    x = cd[["featureA_score","featureB_score","featureC_score"]]
    y = cd["true_melanoma_label"]

    #data_split
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=42)

    #model
    model = LogisticRegression(class_weight="balanced")
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)

    #test the model;
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print(classification_report(y_test, y_pred))
```
